```python
# Dependencies
import pandas as pd
```

```python
# Read csv file
file = "Resources/purchase_data.csv"
pymoli_df = pd.read_csv(file)
```

```python
# Create DataFrame with csv file
pymoli_df.head()
```

```python
# Homework Summary - Total Player Count
totplayercount = pymoli_df["SN"].nunique()

tot_player_count = [{"Total Player Count": totplayercount}]
player_count = pd.DataFrame(tot_player_count)
player_count
```

```python
# Create DataFrame to Summarize Purchasing Analysis, Sales Analysis
sales_analysis_df = pymoli_df[["Purchase ID","Item ID", "Item Name", "Price"]]

# Calculate, unique items, average price, total purchases and total revenue based on the sales_analysis_df
unique_items = sales_analysis_df["Item Name"].nunique()
average_price = sales_analysis_df["Price"].mean()
avg_price = round((average_price), 2)
tot_purchases = sales_analysis_df["Purchase ID"].count()
tot_revenue = sales_analysis_df["Price"].sum()
```

```python
# Homework Summary: Total Purchasing Analysis - Create a summary DataFrame to report results
tot_analysis_summary = pd.DataFrame({"Total Unique Items": [unique_items],
                                    "Average Price": [avg_price],
                                    "Total Purchase": [tot_purchases],
                                    "Total Revenue": [tot_revenue]
                                    })
tot_analysis_summary
```

```python
# Gender Demographichs - Create DataFrame to analyze gender demographics
gender_df = pymoli_df

# Gender Demographics: Get Total Gender Counts and Percentages from the gender_df DataFrame

# Get the Male only count from the gender_df DataFrame
male_count = gender_df.loc[gender_df["Gender"] == "Male"]
male_only = male_count.groupby(["SN"])

# Get the Female only count from the the gender_df DataFrame
female_count = gender_df.loc[gender_df["Gender"] == "Female"]
female_only = female_count.groupby(["SN"])

# Get the Other / Non-Disclosed count from the gender_df DataFrame
other_count = gender_df.loc[gender_df["Gender"] == "Other / Non-Disclosed"]
other_only = other_count.groupby(["SN"])

# Add results from the gender counts
totgender = len(male_only) + len(female_only) + len(other_only)

# Gender Demographics: Calculate the Percentage of Players by gender
m_percent = round((len(male_only) / totgender) *100, 2)
f_percent = round((len(female_only) / totgender) *100, 2)
o_percent = round((len(other_only) / totgender) *100, 2)

m_summary = str(len(male_only))
f_summary = str(len(female_only))
o_summary = str(len(other_only))
```

```python
# Homework Summary - Gender Demographics
gender_data = ({"Total Count": [m_summary, f_summary, o_summary],
               "Percent of Players": [m_percent, f_percent, o_percent]
               })

labels = ['Male', 'Female', 'Other']
gender_demo_summary = pd.DataFrame(gender_data, index=labels)

gender_demo_summary
```

```python
# Purchasing Analysis by Gender - Purchase Count

# Purchase by Male
purchase_by_m = gender_df.loc[gender_df["Gender"] == "Male"]
m_purchase = purchase_by_m.groupby(["Purchase ID"])

# Purchase by Female
purchase_by_f = gender_df.loc[gender_df["Gender"] == "Female"]
f_purchase = purchase_by_f.groupby(["Purchase ID"])

# Purchase by Other / Non-Disclosed
purchase_by_o = gender_df.loc[gender_df["Gender"] == "Other / Non-Disclosed"]
o_purchase = purchase_by_o.groupby(["Purchase ID"])
```

```python
# Purchasing Analysis by Gender - Average Purchase Price

# Columns for calculations

columns = ["SN", "Gender", "Price"]

# Total Purchase and Average Price by Gender
m_sum_price = gender_df.loc[gender_df["Gender"] == "Male", columns]
f_sum_price = gender_df.loc[gender_df["Gender"] == "Female", columns]
o_sum_price = gender_df.loc[gender_df["Gender"] == "Other / Non-Disclosed", columns]

# Total Purchase Value by Gender
m_sum_price["Price"].sum()
f_sum_price["Price"].sum()
o_sum_price["Price"].sum()

# Average Purchase Price by Gender
m_avg_price = gender_df.loc[gender_df["Gender"] == "Male", columns]
f_avg_price = gender_df.loc[gender_df["Gender"] == "Female", columns]
o_avg_price = gender_df.loc[gender_df["Gender"] == "Other / Non-Disclosed", columns]

m_avg = m_avg_price["Price"].mean()
f_avg = f_avg_price["Price"].mean()
o_avg = o_avg_price["Price"].mean()

m_avg_sum = round(m_avg, 2)
f_avg_sum = round(f_avg, 2)
o_avg_sum = round(o_avg, 2)

# Average Total Purchase per Person by Gender
m_avg_tot_pp = round(m_sum_price["Price"].sum() / len(male_only), 2)
f_avg_tot_pp = round(f_sum_price["Price"].sum() / len(female_only), 2)
o_avg_tot_pp = round(o_sum_price["Price"].sum() / len(other_only), 2)
```

```python
# Homework Summary - Purchasing Analysis (Gender)
gender_purchase_data = ({"Purchase Count": [len(m_purchase), len(f_purchase), len(o_purchase)],
              "Average Purchase Price": [m_avg_sum, f_avg_sum, o_avg_sum],
               "Total Purchase Value": [m_sum_price["Price"].sum(), f_sum_price["Price"].sum(), o_sum_price["Price"].sum()],
               "Avg Total Purchase per Person": [m_avg_tot_pp, f_avg_tot_pp, o_avg_tot_pp]
              })

gender_purchase_summary = pd.DataFrame(gender_purchase_data, index=labels)
gender_purchase_summary
```

```python
# Age Demographics
# Create DataFrame for Age Demographics
age_demo_df = pymoli_df

# Create bin for Age Range Column
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
group_names = [" < 10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_demo_df["Age Range"] = pd.cut(age_demo_df["Age"], bins, labels=group_names, include_lowest=True)

# Create Age Range Summary of Age Totals and Age Percentages
age_count_df = age_demo_df.groupby(["Age Range"])
tot_age_count = age_count_df["SN"].nunique()
tot_age_percent = round((tot_age_count / totplayercount) * 100, 2)
```

```python
# Homework Summary - Age Demographics
age_demo_summary = pd.DataFrame({"Total Count": tot_age_count,
                                "Percenatage of Players": tot_age_percent
                                })
age_demo_summary
```

```python
# Age Purchasing Analysis by Age
# Use the age_count_df DataFrame

age_purchase_count = age_count_df["Purchase ID"].nunique()
age_avg_purchase_price = round(age_count_df["Price"].mean(), 2)
age_tot_purchase_value = age_count_df["Price"].sum()
age_tot_per_person = round((age_tot_purchase_value / tot_age_count), 2)


# Homework Summary - Purchasing Analysis by Age

age_purchase_summary = pd.DataFrame({"Purchase Count": age_purchase_count,
                                "Average Purchase Price": age_avg_purchase_price,
                                    "Total Purchase Value": age_tot_purchase_value,
                                    "Avg Total Purchase per Person": age_tot_per_person
                                    })
age_purchase_summary
```

```python
# Top Spenders Analysis
# Create Top Spenders DataFrame

top_spenders = pymoli_df

# Calculate Purchase Count, Average Purchase Price, Total Purchase Value

top_spend_sn = top_spenders.groupby(["SN"])

# Calculations
top_spend_count = top_spend_sn["Purchase ID"].count()
top_spend_avg = round(top_spend_sn["Price"].mean(), 2)
top_spend_value = top_spend_sn["Price"].sum()


# Homework Summary - Top Spenders Summary
top_spenders_summary = pd.DataFrame({"Purchase Count": top_spend_count,
                                "Average Purchase Price": top_spend_avg,
                                    "Total Purchase Value": top_spend_value,
                                    })

top_spenders_summary = top_spenders_summary.sort_values(["Total Purchase Value"], ascending=False)
top_spenders_summary.head()
```

```python
# Most Popular Items Analysis

popular_items = pymoli_df

# Group by Item ID and Item Name
pop_items = popular_items.groupby(["Item ID", "Item Name"])

# Calculations
pop_items_count = pop_items["Item ID"].count()
pop_items_avgprice = round(pop_items["Price"].mean(), 2)
pop_items_value = pop_items["Price"].sum()


# Homework Summary - Most Popular Items Summary
pop_items_summary = pd.DataFrame({"Purchase Count": pop_items_count,
                                "Item Price": pop_items_avgprice,
                                    "Total Purchase Value": pop_items_value,
                                    })

pop_items_summary = pop_items_summary.sort_values(["Purchase Count"], ascending=False)
pop_items_summary.head()
```

```python
# Homework Summary - Most Profitable Items Summary
profit_items_summary = pd.DataFrame({"Purchase Count": pop_items_count,
                                "Item Price": pop_items_avgprice,
                                    "Total Purchase Value": pop_items_value,
                                    })

profit_items_summary = profit_items_summary.sort_values(["Total Purchase Value"], ascending=False)
profit_items_summary.head()
```
